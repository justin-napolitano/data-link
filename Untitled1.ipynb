```python
import nasdaqdatalink as link
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
def read_key():
    link.read_key(".key")
    return True
```

```python
def get_data(ticker):
    data = link.get(ticker)
    return data
```

```python
def main():
    api_link = read_key()
    if api_link:
        ticker = "FED/SUBLPDMHS_XWB_N_Q"
        data = get_data(ticker)
        print(data)
```

```python
class TighteningPipeline():
    def __init__(self):
        self.api_link = self.read_key()
        if self.api_link:
            self.fig, self.ax=self.create_plot()
            self.StandardDF = self.PercentofBanksTighteningStandards()
            self.LoanDemandDF = self.loan_demand()
            self.plot_standards_df()
            #plt.show()
            self.plot_demand_df()
            self.plot_extras()
            plt.show()
            
            #figure = self.plot_demand_df(figure)
            
            
            
    def read_key(self):
        link.read_key(".key")
        return True

    def PercentofBanksTighteningStandards(self):
        ticker = "FED/SUBLPDMHS_XWB_N_Q"
        data = get_data(ticker)
        data = pd.DataFrame(data)
        print('******Standards********')
        print(data)
        return data
    
    def loan_demand(self):
        ticker = "FED/SUBLPDMHD_XWB_N_Q"
        data = get_data(ticker)
        data = pd.DataFrame(data)
        print('******Demand********')
        print(data)
        return data
        
    
    def plot_demand_df(self):
        #plt.style.use('_mpl-gallery')
        x = self.LoanDemandDF.index
        y = self.LoanDemandDF["Value"]
        
        self.ax.plot(x,y, linewidth=3.0, label="Consumer Demand")
        #plt.show()
        #return plt
        
    def plot_standards_df(self):
        #plt.style.use('_mpl-gallery')
        x = self.StandardDF.index
        y = self.StandardDF["Value"]
        
        self.ax.plot(x, y, linewidth=3.0, label="Loan Standards")
        #plt.show()
        
        #return plt
        
    def create_plot(self):
        fig, ax = plt.subplots(figsize=(10,10))
        return fig, ax
    
    def plot_extras(self):
        plt.legend(loc='best')
```

```python
tightening_output = TighteningPipeline()
```
